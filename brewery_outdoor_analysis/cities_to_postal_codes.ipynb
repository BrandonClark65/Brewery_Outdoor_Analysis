```python
import pandas as pd
from uszipcode import SearchEngine
from database_connection import ENGINE
```

```python
# read from `top_us_cities` table
top_us_cities_query = '''
    SELECT *
    FROM top_us_cities
'''

cities = pd.read_sql_query(
    top_us_cities_query,
    con=ENGINE
)
```

```python
cities = cities.sample(20)
cities
```

```python
search = SearchEngine()
```

```python
def city_details_lookup(city, state):
    city_details_dict = {city + ', ' + state: {
        'zipcode': [],
        'state': []
    res = search.by_city_and_state(city='Chicago', state='Illinois', returns=None)
    for r in res:
        city_details_dict[city + ', ' + state]['zipcode'].append(r.zipcode)
        city_details_dict[city + ', ' + state]['state'].append(r.state)
    return city_details_dict
```

```python
output_df = pd.DataFrame()

for i, row in cities.iterrows():
    output_df.append(city_details_lookup(row['city'], row['states']), ignore_index=True)
```
